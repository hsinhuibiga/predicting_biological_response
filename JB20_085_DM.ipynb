JB20_085_DM 粒線體資料清理

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('JB20_085_DM.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df['GT'],df['AD'],df['DP'],df['GQ'],df['PL'] = df['JB20_085'].str.split(':').str
```

```python
df.head(10)
```

```python
df = df.astype(str) 
```

```python
df['AD'] = [[x.split(',')[0]+',' + y for y in x.split(',')[1:]]  for x in df['AD']]
df['ALT'] = df['ALT'].str.split(',')
s = df.pop('AD').explode().reset_index(drop=True)
df = df.explode('ALT').reset_index(drop=True)
df = df.join(s)
```

```python
df.head(780)
```

```python
df.rename(columns={"#CHROM":"CHROM"},inplace=True)
```

```python
df.head(10)
```

```python
df['AD_first'],df['AD_sec']=df['AD'].str.split(',').str
```

```python
df.head(10)
```

```python
df['AD_sec'] = df['AD_sec'].astype(float)
df['DP'] = df['DP'].astype(float)
df['AF'] = df['AD_sec']/df['DP']
```

```python
df.head(20)
```

```python
df=df.drop(['AD_first'],axis=1)
df=df.drop(['AD_sec'],axis=1)
```

```python
df=df.drop(['FORMAT'],axis=1)
df=df.drop(['JB20_085'],axis=1)
df=df.drop(['FILTER'],axis=1)
```

```python
df.head(10)
```

```python
df=df.reindex(columns=['CHROM','POS','ID','REF','ALT','QUAL','INFO','GT','DP','AD','GQ','PL','AF'])
```
